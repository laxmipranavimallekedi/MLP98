```python
import matplotlib.pyplot as plt
stages = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraning', 'KNN Retrained Prediction']

accuracy =  [0.5075, 0.895, 0.963, 0.721, 0.965]
recall =    [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
precision = [0.5327, 0.9045, 0.9061, 1.0000, 0.8873]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(stages, accuracy, marker='o', label='Accuracy')
plt.plot(stages, recall, marker='o', label='Recall')
plt.plot(stages, precision, marker='o', label='Precision')

plt.title('Model Performance Across Stages')
plt.xlabel('Model Stage')
plt.ylabel('Score')
plt.ylim(0.4, 1.05)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
